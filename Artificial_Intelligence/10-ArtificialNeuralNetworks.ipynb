(From PR8) Read a csv file containing a set of mean normalized data with columns containing feature values and the last column representing output values. Return the data as a matrix of floats.

```python
def read_csv( filename):
    csv_matrix = []
    with open(filename, 'rb') as csvfile:
        csv_reader = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC)
        for row in csv_reader:
            csv_matrix.append(row)
    return csv_matrix
```

(From PR8) Return a list of random values [0,1) of the given length. This will be used to initialize a $\theta$ vector, which represents the coefficients that map a set of features to outputs to in a linear or logistic regression.

```python
def random_list( size):
    return [random() for n in range(size)]
```

(From PR8) Return the dot product of two vectors represented as lists.

```python
def dot(list_0, list_1):
    if len(list_0) != len(list_1):
        return None
    dot_product = 0
    for index in range(len(list_0)):
        dot_product += list_0[index]*list_1[index]
    return dot_product
```

Calculate the estimated output values $y=\frac{1}{1+e^{-z}}$, where $z=\theta\cdot x$ and $x$ represents the inputs, for a set of nodes in the network. This is a generalization of the calculation used in logistic regression, and is used to complete a feed forward step in neural network training.

```python
def network_y_hat(theta_list, x):
    return [1/(1+exp(-1*dot(theta, x))) for theta in theta_list]
```

Calculate the logistic error for the network outputs and return the mean error over the entire data set.

```python
def network_error(data, theta_h, theta_y):
    error = []
    for row in data:
        x = [1]+row[:-1*len(theta_y)] # input vector
        y = row[-1*len(theta_y):] # output vector
        y_h = network_y_hat(theta_h, x) # hidden node output values
        y_hat = network_y_hat(theta_y, [1]+y_h) # output node values
        error.append(-1*sum([(y[i]*log(y_hat[i])+(1-y[i])*log(1-y_hat[i])) for i in range(len(theta_y))])/len(theta_y))
    return sum([error[i]**2 for i in range(len(error))])/len(error)
```

Update $\theta$ values using error derivatives $\delta$ according to the relation $\theta_j=\theta_j+\alpha\delta^ix_j$, where $x_j$ represents the preceding signal. The learning rate $\alpha$ directs how much influence each signal (data point) has on the function. This is a generalization of the calculation used in logistic regression, and is used to complete a step of back propogation in neural network training.

```python
def update_theta(theta_list, alpha, delta, signal):
    for i, theta_vector in enumerate(theta_list):
        for j, theta in enumerate(theta_vector):
            theta_list[i][j] += alpha*delta[i]*signal[j]
```

Create a neural network object class that contains $\theta^h$ vectors for the hidden nodes and $\theta^o$ vectors for the output nodes. The neural network is trained by successively applying logistic regressions on each node in the network. For each set of mean normalized data inputs $x$, a feed forward step is applied on the network to calculate first the hidden node outputs $y^h=\frac{1}{1+e^{-z}}$, where $z=\theta^h\cdot x$, and then the estimated outputs $\hat{y}=\frac{1}{1+e^{-z}}$, where $z=\theta^o\cdot y^h$. A back propogation step is then applied to calculate first the $\delta^o$ derivative errors on the output nodes and then the $\delta^h$ derivative errors on the hidden nodes, which are used to update the $\theta$ vectors.

To initialize the network, set the $\theta$ vectors to random values in the range [0,1]. To calculate the expected output for a given instance $x=\{x_1, x_2, ...\}$, apply the $\theta$ vectors for the logistic regressions over the network. The bias $x_0=1$ is implicit in each node in the network.

```python
class NeuralNetwork(object):
    
    def __init__(self, number_inputs, number_hiddens, number_outputs):
        self.theta_h = [random_list(1+number_inputs) for n in range(number_hiddens)] # initialize hidden node theta vectors
        self.theta_y = [random_list(1+number_hiddens) for n in range(number_outputs)] # initialize output node theta vectors
        self.error = [0] # store error values
    
    def train( self, data='neural_network_regression.csv', epsilon=10**-8, alpha=0.01, debug=True):
        if isinstance(data, str): # if a filename is given, read and store the data into a matrix
            data = read_csv(data)
        self.error.append(network_error(data, self.theta_h, self.theta_y)) # mean squared logistic error
        while (abs(self.error[-1]-self.error[-2]) > epsilon):
            for row in data:
                if debug == True:
                    print "mean squared error: " + str(self.error[-1])
                    print "theta_h: " + str(self.theta_h)
                    print "theta_y: " + str(self.theta_y)
                # feed forward step
                x = [1]+row[:-1*len(self.theta_y)] # input vector
                y = row[-1*len(self.theta_y):] # output vector
                y_h = network_y_hat(self.theta_h, x) # hidden node output values
                y_hat = network_y_hat(self.theta_y, [1]+y_h) # output node values
                #back prop step
                delta_o = [y_hat[i]*(1-y_hat[i])*(y[i]-y_hat[i]) for i in range(len(self.theta_y))] # delta vector for outputs
                delta_h = [y_h[j]*(1-y_h[j])*sum([self.theta_y[i][j+1]*delta_o[i] for i in range(len(self.theta_y))]) for j in range(len(self.theta_h))] # delta vector for hidden nodes
                update_theta(self.theta_h, alpha, delta_h, x) # update hidden node derivative values
                update_theta(self.theta_y, alpha, delta_o, [1]+y_h) # update output derivative values
            self.error.append(network_error(data, self.theta_h, self.theta_y)) # mean squared logistic error
        print "mean squared error: " + str(self.error[-1])
        print "theta_h: " + str(self.theta_h)
        print "theta_y: " + str(self.theta_y)
    
    def output( self, instance):
        y_h = network_y_hat(self.theta_h, [1]+instance)
        return network_y_hat(self.theta_y, [1]+y_h)
```

Generate and train $n$ neural networks with given data and parameters, and return the average of the resulting mean square error curves.

```python
def get_mean_squared_error(number_hiddens, data='neural_network_regression.csv', epsilon=10**-9, alpha=0.005, n=5):
    error = []
    for i in range(n):
        neural_network = NeuralNetwork(2, number_hiddens, 1)
        neural_network.train(data, epsilon, alpha, False)
        if error == []:
            error = neural_network.error[1:]
        else:
            error = map(sum, zip(error, neural_network.error[1:]))
    return [value/n for value in error]
```

I selected my parameters by examining the convergence of the error values. I incrementally lowered the degree of the error change tolerance, $\epsilon$, until it took longer than a few seconds to achieve convergence, while incrementally increasing the respective learning factor, $\alpha$, until the sensitivity was too low for convergence to be achieved. This maximized precision by minimizing $\epsilon$ while maintaining performance by maximizing $\alpha$. As the number of hidden nodes increased, more computations were required to update the $\theta$ values, and convergence became more difficult for the same $\epsilon$.

With fewer hidden nodes, a higher $\alpha$ value could be used, leading to better learning performance. With more than one hidden node, the regression seemed to be getting stuck at different local optimums, leading to varying results for the $\theta$ vectors. To improve the implementation of the gradient descent algorithm, I thought it would be beneficial to use random restarts and store the best value over several attempts.

For the purpose of comparison, I chose to use fixed $\epsilon$ and $\alpha$ values, and to plot the mean squared errors after training each network size 5 times. The model tended to achieve convergence more efficiently as the number of hidden nodes increased from 1 to 3. As discussed in the lectures, one layer of hidden nodes should be sufficient for modeling a continuous function, with each node representing an inflection point in the data.
