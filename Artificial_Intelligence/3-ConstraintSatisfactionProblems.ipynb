Define a global variable containing a list of the four colors in the domain:

```python
domain_4color = ["color1","color2","color3","color4"]
```

Set the domain of each of the nodes to the four colors. This will be used to track the remaining allowable colors for each node.

```python
def get_domain( vertices):
    domain = {}
    for vertex in vertices:
        domain[vertex] = deepcopy(domain_4color)
    return domain
```

Convert a graph represented as a list of vertices and edges (V,E) to a graph represented as an adjacency list. The graph is returned as an unsorted dict of vertices mapped to a sorted list of adjacent vertices.

```python
def adjacency_list( vertices, edges):
    graph = {}
    for vertex in vertices: # add an adjacency list for each vertex
        graph[vertex] = []
        for edge in edges: # convert edges to entries in adjacency lists
            if vertex == edge[0]:
                graph[vertex].append(edge[1])
            elif vertex == edge[1]:
                graph[vertex].append(edge[0])
        graph[vertex].sort() # alphanumerically sort the entries in each adjacency list
    return graph
```

Check if to see whether a graph has been completely colored by checking that the map coloring contains an entry for every node in the graph.

```python
def is_colored( graph, map_coloring):
    for node in graph:
        if node not in map_coloring:
            return False
    return True
```

Check to see if coloring a new node is consistent with the current map colorings by comparing them with any adjacent nodes that have already been colored. Also apply forward checking to ensure that any uncolored nodes that are adjacent to the new node can still be colored differently.

```python
def is_consistent( graph, map_coloring, domain, node, color):
    for adjacent_node in graph[node]: # check all adjacent nodes
        if adjacent_node in map_coloring: # make sure no adjacent nodes are already the same color
            if map_coloring[adjacent_node] == color:
                return False
        elif adjacent_node in domain: # look forward to uncolored adjacent nodes
            if len(domain[adjacent_node]) == 1 and domain[adjacent_node][0] == color: # check each edge (border)
                return False
    return True
```

Get a list of nodes that have yet to be colored in the graph.

```python
def get_uncolored( graph, map_coloring):
    uncolored_nodes = []
    for node in graph:
        if node not in map_coloring:
            uncolored_nodes.append(node)
    return uncolored_nodes
```

Add a colored node. Return the new map colorings. This represents a new state to be checked.

```python
def get_new_coloring( map_coloring, node, color):
    map_coloring_new = map_coloring
    map_coloring_new[node] = color
    return map_coloring_new
```

Add a colored node. Apply forward looking and remove that color from the domain of adjacent nodes. 

```python
def get_new_domain( graph, domain, node, color):
    domain_new = domain
    domain_new[node].remove(color)
    for adjacent_node in graph[node]: # check all adjacent nodes
        if adjacent_node in domain:
            if color in domain_new[adjacent_node]:
                domain_new[adjacent_node].remove(color)
    del domain_new[node] # (optionally) delete node from domain
    return domain_new
```

Apply the Minimum Remaining Values heuristic, which selects the next node based on which node has the least values remaining in its domain, with static ordering as a tiebreaker. When selecting nodes, it is beneficial to start with the most constrained nodes to reduce the amount of backtracking required from failure states.

```python
def get_node_min_remaining( domain, uncolored_nodes):
    node_min_remaining = uncolored_nodes[0]
    for node in uncolored_nodes:
        if len(domain[node]) < len(domain[node_min_remaining]): # find node with smallest domain
            node_min_remaining = node
    return node_min_remaining
```

Apply the Degree Heuristic, which selects the next node based on which node has the greatest number of adjacent nodes (constraints), with static ordering as a tiebreaker. This is not dynamic, which means that previously covered constraints are still counted. This is similar to the concept used in Minimum Remaining Values, but with a different measure used for prioritization.

```python
def get_node_degree_heuristic( graph, uncolored_nodes):
    node_max_degree = uncolored_nodes[0]
    for node in uncolored_nodes:
        if len(graph[node]) > len(graph[node_max_degree]): # find node with largest adjacency (constraint) list
            node_max_degree = node
    return node_max_degree
```

Apply the Least Constraining Value heuristic, which selects the next color to try for a selected node based on which color will constrain the least number of adjacent nodes. When assigning values, it is beneficial to leave the more options open for the remaining uncolored nodes to reduce the chance of arriving at a failure state. 

__Uncommented the print statement below to show a detailed trace of the program operation.__

```python
def get_color_least_constraining( graph, domain, node):
    node_colors = domain[node] # the colors available for the node
    node_color_count = {} # track the number of adjacent nodes that share each posssible node color
    for color in node_colors:
        node_color_count[color] = 0
    for adjacent_node in graph[node]: # search the adjacent nodes
        if adjacent_node in domain:
            for color in node_colors: # search the colors in the domains in each adjacent node
                if color in domain[adjacent_node]:
                    node_color_count[color] += 1
    print "Constraint Count: " + str(node_color_count) #trace
    return sorted(node_colors, key=node_color_count.get) # sort the colors by the number of constraints
```

This is the primary recursive function that implements backtracking to iterate through states generated by coloring nodes one by one. Forward looking, node prioritization, and color selection improve the efficiency of searching for a solution because they reduce the amount of failure states that will be encountered and backtracked from.

The graph of vertices and edges (V,E) is converted to an adjacency list to make it easier to iterate and reference the nodes that are adjacent to each node (the bordering regions on the map).

```python
def get_coloring( graph, map_coloring, domain):
    if is_colored(graph, map_coloring):
        return map_coloring
    else:
        uncolored_nodes = get_uncolored(graph, map_coloring)
        uncolored_nodes.sort()
        #node = uncolored_nodes[0] # select node based on static ordering
        node = get_node_min_remaining(domain, uncolored_nodes) # select node based on Minimum Remaining Values
        #node = get_node_degree_heuristic(graph, uncolored_nodes) # select node based on Degree Heuristic
        #node_colors = domain[node] # select color based on static ordering
        node_colors = get_color_least_constraining(graph, domain, node)
        print "Domain List: " + str(domain) #trace
        print "Uncolored Nodes: " + str(uncolored_nodes) #trace
        print "Selected Node: " + str(node) #trace
        print "Selected Colors: " + str(node_colors) + "\n" #trace
        for color in node_colors: # check all possible states created by coloring the node
            if is_consistent(graph, map_coloring, domain, node, color):
                solution = get_coloring(graph, get_new_coloring(map_coloring, node, color), 
                                        get_new_domain(graph, domain, node, color))
                if solution != None:
                    return solution
        return None # the Four Color Theorem states that this should never be the final result
```

This is the main function that converts the graph representation into an adjacency list and calls the backtracking and forward checking algorithm to solve the map coloring problem. It returns a dictionary of nodes mapped to colors, where no two adjacent nodes have the same color.

```python
def color_map( world):
    map_coloring = {} # store the state containing all colored nodes
    vertices = world[0]
    edges = world[1]
    graph = adjacency_list(vertices, edges) # convert graph (V,E) to an adjacency list representation
    domain = get_domain(vertices)
    print "Adjacency List: " + str(graph) + "\n" #trace
    return get_coloring(graph, map_coloring, domain) # get completed map coloring
```
