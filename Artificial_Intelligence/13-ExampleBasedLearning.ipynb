Read a csv file containing a set of data with the first column containing $y$ labels and the remainder of the columns containing a vector of attribute values $x$. Return the data as a matrix of floats in the form [[$x_1,x_2,...$], [$y$]].

```python
def read_csv( filename):
    csv_matrix = [[],[]]
    with open(filename, 'rb') as csv_file:
        csv_reader = csv.reader(csv_file, quoting=csv.QUOTE_NONNUMERIC)
        for row in csv_reader:
            csv_matrix[0].append(row[1:]) # x values
            csv_matrix[1].append(row[0]) # y values
    csv_file.closed
    return csv_matrix
concrete_data = read_csv('concrete_compressive_strength.csv')
#print concrete_data #test
```

[From PR11] Shuffle a set of data. This will be used to randomize the data before partitioning it for n-fold validation.

```python
def shuffle_data(data):
    shuffled_data = [[],[]]
    indices = range(len(data[0]))
    shuffle(indices)
    for index in indices:
        shuffled_data[0].append(data[0][index])
        shuffled_data[1].append(data[1][index])
    return shuffled_data
```

[From PR11] Split a set of data into $n$ equal partitions. Each of the sublists will represent a test/training fold.

```python
def get_partitioned_data(data, n):
    partitioned_data = [[[],[]] for i in range(n)]
    for index in range(len(data[0])):
        partitioned_data[index%n][0].append(data[0][index])
        partitioned_data[index%n][1].append(data[1][index])
    return partitioned_data
```

Calculate the Euclidean distance between two points of the same dimension, $x$ and $y$, as $d(x,y)=(\sum\limits_{i}(x_i-y_i)^2)^{1/2}$. This will be used to search for the k nearest neighbors of a point.

```python
def euclidean_distance(x, y):
    distance = 0
    for i in range(len(x)):
        distance += (x[i] - y[i])**2
    return sqrt(distance)
#print euclidean_distance([0,5],[12,0]) #test
```

Retrieve the k-nearest neighbor value using regression, which takes the average of the y values for the k nearest neighbors.

```python
def knn_regression(knn_values):
    return sum(knn_values)/len(knn_values)
#print knn_regression([3.0,2.0,3.0,2.0,3.0]) #test
```

Retrieve the k-nearest neighbor value using classification, which takes the majority y value for the k nearest neighbors.

```python
def knn_classification(knn_values):
    majority_value = None
    max_count = -1*float('inf')
    counts = {}
    for value in knn_values:
        if value in counts:
            counts[value] += 1
        else:
            counts[value] = 1
    for value, count in counts.items():
        if count > max_count:
            max_count = count
            majority_value = value
    return majority_value
#print knn_classification([3.0,2.0,3.0,2.0,3.0]) #test
```

Use k-nearest neighbor to estimate a value from a dataset. Given a query, find the k points with the smallest Euclidean distance from the query. Calculating the distances is done in $O(n)$ time and storing the k points with the least distances is done in $O(k\log k)$ time. Then return the value estimated from these k points using either regression or classification.

```python
def knn(k, data, query, regression=True):
    knn_y = data[1][:k] # store sorted array of nearest neighboring values
    knn_d = [euclidean_distance(x, query) for x in data[0][:k]] # store sorted array of nearest neighboring distances
    knn_y, knn_d = (list(i) for i in zip(*sorted(zip(knn_d, knn_y)))) # sort both arrays according to the distances
    for i in range(k, len(data[0])): # for each point in the dataset
        d = euclidean_distance(data[0][i], query) # calculate the Euclidean distance
        if d < knn_d[-1]: # if a closer point is found, replace the farthest neighbor with the point
            knn_y[-1] = data[1][i]
            knn_d[-1] = d
            knn_y, knn_d = (list(i) for i in zip(*sorted(zip(knn_d, knn_y)))) # sort both arrays according to the distances
    if regression == True:
        return knn_regression(knn_y)
    else:
        return knn_classification(knn_y)
```

Apply k-nearest neighbor on the given training data to and return the mean absolute error over each of the test values. This will be used to plot learning curves and evaluate performance.

```python
def get_error(test_data, train_data, k, debug=False):
    y = test_data[1]
    y_hat = [knn(k, train_data, query) for query in test_data[0]]
    error = sum(abs(y[i]-y_hat[i])/y[i] for i in range(len(y)))/len(y)
    if debug == True:
        print "error: " + str(error)
    return error
```

Incrementally increase the amount of data points used for k-nearest neighbor to produce an error learning curve for the given training and test data, and for the specified k value.

```python
def get_learning_curve(test_data, train_data, k, training_rate=0.2):
    number_trained = [] # number of data instances used at each training step
    test_error = [] # test data learning curve
    train_error = [] # training data learning curve
    n_train = int(training_rate*len(train_data[0]))
    while n_train < int((1+training_rate)*len(train_data[0])):
        test_error.append(get_error([test_data[0], test_data[1]], 
                                    [train_data[0][:n_train], train_data[1][:n_train]], k)) # error for test data
        train_error.append(get_error([train_data[0][:n_train], train_data[1][:n_train]], \
                                     [train_data[0][:n_train], train_data[1][:n_train]], k)) # error for training data
        number_trained.append(n_train) # size of data trained at this step
        n_train += int(training_rate*len(train_data[0]))
    return [number_trained, test_error, train_error]
```

Partition the given data into n test/training folds to produce an averaged learning curve for the specified k value.

```python
def get_averaged_learning_curves(data, k, n=3):
    data = shuffle_data(data) # shuffle the data in case it is ordered
    partitioned_data = get_partitioned_data(data, n) # partition the data into n folds
    learning_curves = []
    for i in range(n):
        print "k=" + str(k) + ": Fold " + str(i)
        test_data = partitioned_data[i] # select a test fold
        train_data_x = [x for fold in partitioned_data[:i]+partitioned_data[i+1:] for x in fold[0]] # select a training fold
        train_data_y = [y for fold in partitioned_data[:i]+partitioned_data[i+1:] for y in fold[1]] # select a training fold
        learning_curves.append(get_learning_curve(test_data, [train_data_x, train_data_y], k)) # performance metrics
    return [[sum(learning_curves[i][index][j] for i in range(len(learning_curves)))/len(learning_curves) \
             for j in range(len(learning_curves[0][index]))] for index in range(len(learning_curves[0]))]
```

Plot learning curves for k-nearest neighbor with for k = m,m+1,...,n-1,n. This will be used to determine the optimal value of k for a data set.

```python
def plot_learning_curves(data, m=32, n=39, step=1):
    plt.title('k-Nearest Neighbor Test Error')
    plt.ylabel('Mean Error')
    plt.xlabel('Size of Training Data')
    for k in range(m,n+1,step):
        print "plotting " + str(k) + "-nearest neighbors..."
        learning_curves = get_averaged_learning_curves(data, k)
        plt.plot(learning_curves[0], learning_curves[1], label=str(k)+"test")
        plt.plot(learning_curves[0], learning_curves[2], label=str(k)+"train")
    plt.legend(loc='upper right')
    plt.show() # display plots
```

Perform n-fold (10-fold by default) cross-validation on a set of data, which allows the performance of a model to be tested on a fixed set of data. Shuffle and partition the data in $n$ equal folds. Iterate through each of the $n$ folds, designating the selected partition as the test data and the remaining partitions as the training data for k-nearest neighbor.

```python
def n_fold(data, n=10, k=34, debug=False):
    metrics = [] # store of data metrics for each test fold
    if n == 1:
        metrics.append(get_error(data, data, k)) # performance metrics
    else:
        data = shuffle_data(data) # shuffle the data in case it is ordered
        partitioned_data = get_partitioned_data(data, n) # partition the data into n folds
        for i in range(n):
            print "Fold: " + str(i)
            test_data = partitioned_data[i] # select a test fold
            train_data_x = [x for fold in partitioned_data[:i]+partitioned_data[i+1:] for x in fold[0]] # select a training fold
            train_data_y = [y for fold in partitioned_data[:i]+partitioned_data[i+1:] for y in fold[1]] # select a training fold
            metrics.append(get_error(test_data, [train_data_x, train_data_y], k, True)) # performance metrics
    mean = sum(metrics[i] for i in range(n))/n # mean error
    std_dev = sqrt(sum((metrics[i]-mean)**2 for i in range(n))/n) # standard deviation of the mean
    std_err = std_dev/sqrt(n) # standard error of the mean
    print str(n) + "-fold Mean Error: " + str(mean)
    print str(n) + "-fold Standard Deviation of the Mean: " + str(std_dev)
    print str(n) + "-fold Standard Error of the Mean: " + str(std_err)
    print str(n) + "-fold 95% Confidence Interval: [" + str(mean-2.228*std_err) + "," + str(mean+2.228*std_err) + "]"
```
