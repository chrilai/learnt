Shuffle a set of data. This will be used to randomize the data before partitioning it for n-fold validation.

```python
def shuffle_data(data):
    shuffled_data = [[],[]]
    indices = range(len(data[0]))
    shuffle(indices)
    for index in indices:
        shuffled_data[0].append(data[0][index])
        shuffled_data[1].append(data[1][index])
    return shuffled_data
```

Split a set of data into $n$ equal partitions. Each of the sublists will represent a test/training fold.

```python
def get_partitioned_data(data, n):
    partitioned_data = [[[],[]] for i in range(n)]
    for index in range(len(data[0])):
        partitioned_data[index%n][0].append(data[0][index])
        partitioned_data[index%n][1].append(data[1][index])
    return partitioned_data
```

Compare the predicted labels $\hat{y}$ against the actual labels $y$. Count the number of true and false positives and negatives, and return them in a 2x2 matrix [[TP, FP], [FN, TN]].

```python
def get_confusion_matrix(y, y_hat, positive):
    confusion_matrix = [[0, 0],[0, 0]]
    for index in range(len(y)):
        if y_hat[index] == positive: # predicted value positive
            if y[index] == positive: # true positive
                confusion_matrix[0][0] += 1
            else: # false positive
                confusion_matrix[0][1] += 1
        else: # predicted value negative
            if y[index] == positive: # false negative
                confusion_matrix[1][0] += 1
            else: # true negative
                confusion_matrix[1][1] += 1
    return confusion_matrix
```

(From PR9) Calculate and return the accuracy using confusion matrix values $\frac{TP+TN}{n}$. This represents the proportion of total values that were correctly predicted.

```python
def get_accuracy(conf_matrix):
    return float(conf_matrix[0][0]+conf_matrix[1][1])/(conf_matrix[0][0]+conf_matrix[0][1]+conf_matrix[1][0]+conf_matrix[1][1])
```

(From PR9) Calculate and return the error using confusion matrix values $\frac{FP+FN}{n}=1-\frac{TP+TN}{n}$. This represents the proportion of total values that were incorrectly predicted.

```python
def get_error(conf_matrix):
    return 1.0-get_accuracy(conf_matrix)
```

(From PR9) Calculate and return the precision using confusion matrix values $\frac{TP}{TP+FP}$. This represents the proportion of the predicted positives that were correctly predicted.

```python
def get_precision(conf_matrix):
    if (conf_matrix[0][0]+conf_matrix[0][1]) == 0:
        return 1.0
    return float(conf_matrix[0][0])/(conf_matrix[0][0]+conf_matrix[0][1])
```

(From PR9) Calculate and return the recall using confusion matrix values $\frac{TP}{TP+FN}$. This represents the proportion of the actual positives that were correctly predicted.

```python
def get_recall(conf_matrix):
    if (conf_matrix[0][0]+conf_matrix[1][0]) == 0:
        return 1.0
    return float(conf_matrix[0][0])/(conf_matrix[0][0]+conf_matrix[1][0])
```

Read a csv file containing a set of data with the first column containing $y$ labels and the remainder of the columns containing a vector of attribute values $x$. Return the data as a matrix of the form [[$x_1,x_2,...$], [$y$]]. Missing values are marked as '?'.

```python
def read_csv( filename):
    csv_matrix = [[],[]]
    with open(filename, 'rb') as csv_file:
        csv_reader = csv.reader(csv_file)
        for row in csv_reader:
            csv_matrix[0].append(row[1:]) # x values
            csv_matrix[1].append(row[0]) # y values
    return csv_matrix
#data = read_csv('agaricus-lepiota.data')
#print data #test
```

Read a file containing a numbered list of attributes and values codes. The first line contains the class labels. Each line number/attribute is followed by a '.', each attribute name is followed by a ':', and each attribute name=character code mapping is separated by a ','.

```python
def read_definitions( filename):
    definitions = {}
    with open(filename, 'rb') as def_file:
        for row in def_file:
            line = "".join(row.split()) # remove whitespace
            line = line.split('.')
            attribute_num = int(line[0]) - 1 # attribute number
            line = line[1].split(':')
            attribute = line[0] # attribute
            definitions[attribute_num] = attribute # add attributes to dict
            names = line[1].split(',')
            attribute_dict = {} # dict of attribute values
            for name in names:
                line = name.split('=')
                attribute_value = line[0]
                attribute_char = line[1]
                attribute_dict[attribute_char] = attribute_value
            definitions[attribute] = attribute_dict # add values to dict
    return definitions
#definitions = read_definitions('agaricus-lepiota.names.data')
#print definitions #test
```

Create a Node class to represent the nodes in a decision tree. Each node stores an attribute value, a class label, and a list of child nodes.

```python
class Node(object):
    def __init__(self, size=0, label=None, next_attribute=None, attribute=-1, value=None):
        self.size = size # number of data points at this stage of decision tree
        self.label = label # label
        self.next_attribute = next_attribute # attribute
        self.attribute = attribute # attribute
        self.value = value # attribute value
        self.children = [] # child node array
    
    def __repr__(self):
        if self.children == []:
            return str(self.value) + ':' + str(self.label)
        return str(self.attribute) + ':' + str(self.value)
    
    def reset(self, size=0):
        self.size = size
        self.label = None
        self.next_attribute = None
        self.attribute = -1
        self.value = None
        self.children = []
```

Check if a set of data is homogeneous, that is, whether every data point has the same label. This is used for abase case in the ID3 algorithm.

```python
def is_homogeneous(data):
    first_label = data[1][0]
    for label in data[1][1:]:
        if label != first_label:
            return False
    return True
#print is_homogeneous(data) #test
```

Find and return the majority label for a set of data. This is used for a base case and default value in the ID3 algorithm.

```python
def majority_label(data):
    majority_label = None
    label_counts = {}
    max_label_count = 0
    for label in data[1]:
        if label not in label_counts:
            label_counts[label] = 1
        else:
            label_counts[label] += 1
    for label in label_counts:
        if label_counts[label] > max_label_count:
            max_label_count = label_counts[label]
            majority_label = label
    return majority_label
#print majority_label(data) #test
```

Split a set of data based on the values of an attribute and map resulting subsets to a dict. This is used to produce a set of child nodes on the decision tree.

```python
def get_split_data(data, attribute):
    split_data = {}
    for index, x in enumerate(data[0]):
        if x[attribute] not in split_data:
            split_data[x[attribute]] = [[data[0][index]],[data[1][index]]]
        else:
            split_data[x[attribute]][0].append(data[0][index])
            split_data[x[attribute]][1].append(data[1][index])
    return split_data
#print get_split_data(data, 0) #test
```

Calculate the entropy, $E(S)=-\sum\limits_{i}{p_ilog_2(p_i)}$, of a set of data with labels having a $p_i$ probability of occurrence.

```python
def get_entropy(data):
    entropy = 0
    label_counts = {}
    for label in data[1]:
        if label not in label_counts:
            label_counts[label] = 1
        else:
            label_counts[label] += 1
    for label in label_counts:
        p_label = float(label_counts[label])/len(data[1])
        entropy += -1*p_label*log2(p_label)
    return entropy
#print get_entropy(data)
```

Select the best attribute with which to split a set of data by calculating the entropy of each subset, $E(S_v)$, and finding the attribute, $A$, that leads to the most information gain, $G(S,A)=E(S)-\sum\limits_{v\in V_A}\frac{|S_v|}{|S|}E(S_v)$.

```python
def get_best_attribute(data, attributes):
    best_attribute = None
    max_information_gain = -1*float('inf')
    for attribute in attributes:
        information_gain = get_entropy(data)
        split_data = get_split_data(data, attribute)
        for value in split_data:
            information_gain += -1*(len(data)/len(split_data[value]))*get_entropy(split_data[value])
        if information_gain > max_information_gain:
            max_information_gain = information_gain
            best_attribute = attribute
    return best_attribute
#print get_best_attribute(data, range(len(definitions)/2 - 1)) #test
```

Return a copy of a list with the specified item removed. This will be used to update the list of attributes.

```python
def remove_list(list, item):
    new_list = deepcopy(list)
    new_list.remove(item)
    return new_list
```

Return the largest child node of a specified node. This is used as the best guess to determine the correct label for a data instance that contains an attribute value that is not present in the decision tree.

```python
def get_largest_child(node):
    max_size = 0
    largest_child = None
    for child in node.children:
        if child.size > max_size:
            max_size = child.size
            largest_child = child
    return largest_child
```

Perform the ID3 algorithm to generate a decision tree using a set of data. With each iteration, we select the best attribute to split the data into subsets so that the entropy of the data labels is minimized, with each subset being represented as a new child node. Stopping cases occur at the leaf nodes when the data subset is empty or homogeneous, or no more attributes remain to be processed. 

```python
def id3(data, attributes, default):
    if data == []: # if data is empty, stop and return current node
        return Node(len(data[0]), default)
    if is_homogeneous(data): # if data is homogeneous, return current node with the class label
        return Node(len(data[0]), data[1][0])
    if attributes == []: # if the attributes are empty, return current node with the majority label
        return Node(len(data[0]), majority_label(data))
    best_attribute = get_best_attribute(data, attributes)
    split_data = get_split_data(data, best_attribute)
    split_attributes = remove_list(attributes, best_attribute)
    node = Node(len(data[0]), default, best_attribute)
    for value in split_data: # create child node with attribute value and majority label
        child = id3(split_data[value], split_attributes, majority_label(data))
        child.attribute = best_attribute
        child.value = value
        node.children.append(child)
    return node
```

Create a class to represent a decision tree model. The init function initiates the decision tree by instantiating a root node and processing a set of data definitions. The train function processes a set of data examples and trains the decision tree using the ID3 algorithm. For data with missing values ('?'), the missing values are treated as implicit attribute values.

The output function uses the decision tree to guess the correct label for a given instance of data. If an attribute value is present in the instance but not in the tree, select the largest child node as the best guess for the attribute value.

```python
class DecisionTree(object):
    
    def __init__(self, definitions='agaricus-lepiota.names.data'):
        if isinstance(definitions, str): # if a filename is given, read and store the definitions into a dict
            definitions = read_definitions(definitions)
        self.root = Node()
        self.num_attributes = len(definitions)/2 - 1
        self.definitions = definitions
    
    def train(self, data='agaricus-lepiota.data'):
        if isinstance(data, str): # if a filename is given, read and store the data into a list of lists
            data = read_csv(data)
        attributes = range(self.num_attributes)
        self.root.reset(len(data[0])) # reset root node
        self.root = id3(data, attributes, majority_label(data)) # carry out ID3 algorithm to train decision tree
        
    def output(self, instance):
        node = self.root # start with root node
        while(node.children != []): # until a leaf node is reached
            match_found = False
            for child in node.children: # search child nodes
                if instance[child.attribute] == child.value: # if a matching value for the attribute is found
                    node = child
                    match_found = True
                    break
            if (match_found == False): # if a matching value for the attribute is not fond
                node = get_largest_child(node) # default to largest child
        return node.label
```

Return a string representation of a given node in a tree. Return 'root' for the root node, the attribute and value for intermediate nodes, and the attribute, value, and label for leaf nodes.

```python
def print_node(tree, node):
    if node.attribute == -1:
        return 'root'
    attribute_str = tree.definitions[node.attribute]
    value_str = tree.definitions[attribute_str][node.value]
    label_str = tree.definitions[tree.definitions[-1]][node.label]
    if node.children == []:
        return attribute_str + ':' + value_str + ':' + label_str
    return attribute_str + ':' + value_str
```

Add the nodes of a tree to the given graph, and add edges between each node and its children.

```python
def graph_tree(graph, tree, node, positive='p'):
    if print_node(tree, node) == 'root':
        graph.add_node(print_node(tree, node))
    for child in node.children:
        graph.add_node(print_node(tree, child))
        graph.add_edge(print_node(tree, node), print_node(tree, child))
        graph_tree(graph, tree, child)
```

Print the given tree starting from the given root node.

```python
def print_tree(tree, node):
    G = nx.Graph()
    graph_tree(G, tree, node)
    nx.draw(G)
    plt.show()
```

Train the given data to produce a decision tree and calculate the resulting accuracy/error, precision, and recall. Return a list containing the resulting data metrics.

```python
def get_metrics(test_data, train_data, debug=False, positive='p'):
    decision_tree = DecisionTree()
    decision_tree.train(train_data)
    y = test_data[1]
    y_hat = [decision_tree.output(x) for x in test_data[0]]
    confusion_matrix = get_confusion_matrix(y, y_hat, positive)
    accuracy = get_accuracy(confusion_matrix)
    precision = get_precision(confusion_matrix)
    recall = get_recall(confusion_matrix)
    print "confusion matrix: " + str(confusion_matrix)
    if debug == True:
        print "accuracy: " + str(accuracy)
        print "error: " + str(1-accuracy)
        print "precision: " + str(precision)
        print "recall: " + str(recall)
    return [accuracy, precision, recall]
```

Perform n-fold (10-fold by default) cross-validation on a set of data, which allows the performance of a model to be tested on a fixed set of data. Shuffle and partition the data in $n$ equal folds. Iterate through each of the $n$ folds, designating the selected partition as the test data and the remaining partitions as the training data for logistic regression.

```python
def n_fold(n=10, data='agaricus-lepiota.data', debug=False):
    if isinstance(data, str): # if a filename is given, read and store the data into a matrix
        data = read_csv(data)
    metrics = [] # store of data metrics for each test fold
    if n == 1:
        metrics.append(get_metrics(data, data)) # performance metrics
    else:
        data = shuffle_data(data) # shuffle the data in case it is ordered
        partitioned_data = get_partitioned_data(data, n) # partition the data into n folds
        for i in range(n):
            print "Fold: " + str(i)
            test_data = partitioned_data[i] # select a test fold
            train_data_x = [x for fold in partitioned_data[:i]+partitioned_data[i+1:] for x in fold[0]] # select a training fold
            train_data_y = [y for fold in partitioned_data[:i]+partitioned_data[i+1:] for y in fold[1]] # select a training fold
            metrics.append(get_metrics(test_data, [train_data_x, train_data_y])) # performance metrics
    print str(n) + "-fold Average Accuracy: " + str(sum(metrics[i][0] for i in range(n))/n)
    print str(n) + "-fold Average Error: " + str(sum(1-metrics[i][0] for i in range(n))/n)
    print str(n) + "-fold Average Precision: " + str(sum(metrics[i][1] for i in range(n))/n)
    print str(n) + "-fold Average Recall: " + str(sum(metrics[i][2] for i in range(n))/n)
```
