Read a csv file containing a set of mean normalized data with columns containing feature values and the last column representing output values. Return the data as a matrix of floats.

```python
def read_csv( filename):
    csv_matrix = []
    with open(filename, 'rb') as csvfile:
        csv_reader = csv.reader(csvfile, quoting=csv.QUOTE_NONNUMERIC)
        for row in csv_reader:
            csv_matrix.append(row)
    return csv_matrix

print read_csv('linear_regression.csv') #test
```

Return a list of random values of the given length. This will be used to initialize a $\theta$ vector, which represents the coefficients that map a set of features to outputs to in a linear or logistic regression.

```python
def random_list( size):
    return [random.random() for n in range(size)]

print random_list(3) #test
```

Return the dot product of two vectors represented as lists.

```python
def dot(list_0, list_1):
    if len(list_0) != len(list_1):
        return None
    dot_product = 0
    for index in range(len(list_0)):
        dot_product += list_0[index]*list_1[index]
    return dot_product

print dot([1,2,3], [4,5,6]) #test
```

Calculate and return the estimated output values for linear regression, $\hat{y}=\theta\cdot x$.

```python
def linear_y_hat( theta, data):
    return [dot(theta, [1]+row[:-1]) for row in data]
```

Calculate and return the estimated output values for logistic regression, $\hat{y}=\frac{1}{1+e^{-\theta\cdot x}}\rightarrow\{0|1\}$.

```python
def logistic_y_hat( theta, data):
    y_hat = [1/(1+math.exp(-1*dot(theta, [1]+row[:-1]))) for row in data]
    return y_hat
```

Calculate and return the error cost function for linear regression, $J(\theta)=\frac{1}{2n}\sum\limits_{i=1}^n{(\hat{y}^{(i)}-y^{(i)})^2}$, where $\hat{y}=\theta\cdot x$ are estimated output values and $y$ are actual output values. The estimated output values are generated from the dot product of the $\theta$ and $x$ vectors.

```python
def linear_error( data, y_hat):
    return sum((y_hat[i]-data[i][-1])**2 for i in range(len(data)))/(2*len(data))
```

Calculate and return the error cost function for logistic regression, $J(\theta)=-\frac{1}{n}\sum\limits_{i=1}^n{[y^{(i)}\log\hat{y}^{(i)}+(1-y^{(i)})\log(1-\hat{y}^{(i)})]}$, where $\hat{y}=\frac{1}{1+e^{\theta\cdot x}}\rightarrow\{0|1\}$ are estimated output values and $y$ are actual output values. The estimated output values are generated from the dot product of the $\theta$ and $x$ vectors.

```python
def logistic_error( data, y_hat):
    return -1*sum((data[i][-1]*math.log(y_hat[i])+(1-data[i][-1])*math.log(1-y_hat[i])) for i in range(len(data)))/len(data)
```

Calculate and return a vector of partial derivatives of the error cost function for regression, $\frac{\partial J}{\partial\theta_{j}}=\frac{1}{n}\sum\limits_{i=1}^n{(\hat{y}^{(i)}-y^{(i)})x^{(i)}}$. This will be used to update the $\theta$ vector.

```python
def derivative( theta, data, y_hat):
    d_theta = []
    for j in range(len(theta)):   
        if j == 0: # x_0 = 1
            derivative = sum((y_hat[i]-data[i][-1]) for i in range(len(data))) # sum of all error slopes
        else:
            derivative = sum((y_hat[i]-data[i][-1])*data[i][j-1] for i in range(len(data))) # sum of all error slopes
        derivative = derivative/(len(data)) # average values
        d_theta.append(derivative)
    return d_theta
```

Update the values in the $\theta$ vector $\theta_{j}=\theta_{j}-\alpha\frac{\partial J}{\partial\theta_{j}}$. This completes a step of gradient descent. The value of \alpha represents the learning rate, with higher values increasing the rate of convergence but also increasing the risk of overshooting or divergence.

```python
def theta_next( theta, d_theta, alpha):
    for j in range(len(theta)):
        theta[j] = theta[j] - alpha*d_theta[j]
```

Linear regression acts on datasets containing numeric output values. Unique functions are called to determine $\hat{y}$ and $J(\theta)$ to update the $\theta$ vector, which provides a set of values that best estimates the actual $y$ values from $\hat{y}$.

```python
def linear_regression( data, alpha=0.2, epsilon=10**-10, debug=False):
    if isinstance(data, str): # if a filename is given, read and store the data into a matrix
        data = read_csv(data)
    theta = random_list(len(data[0])) # initialize the theta vector to be the same length as the x vectors (including x_0)
    y_hat = linear_y_hat(theta, data) # estimated output values
    error_prev = 0 # initialize previous error to 0
    error = linear_error(data, y_hat) #calculate error
    while (abs(error-error_prev) > epsilon):
        if debug == True:
            print error, theta #trace
        d_theta = derivative(theta, data, y_hat)
        theta_next(theta, d_theta, alpha) # update theta values
        y_hat = linear_y_hat(theta, data) # estimated output values
        error_prev = error # set previous error to current error
        error = linear_error(data, y_hat) # calculate error
    return theta
```

Logistic regression acts on datasets containing categorical output values. Unique functions are called to determine $\hat{y}$ and $J(\theta)$ to update the $\theta$ vector, which provides a set of values that best estimates the actual $y$ values from $\hat{y}$ after a threshold selection function has been applied to set it to 0 or 1.

```python
def logistic_regression( data, alpha=0.2, epsilon=10**-10, debug=False):
    if isinstance(data, str): # if a filename is given, read and store the data into a matrix
        data = read_csv(data)
    theta = random_list(len(data[0])) # initialize the theta vector to be the same length as the x vectors (including x_0)
    y_hat = logistic_y_hat(theta, data) # estimated output values
    error_prev = 0 # initialize previous error to 0
    error = logistic_error(data, y_hat) #calculate error
    while (abs(error-error_prev) > epsilon):
        if debug == True:
            print error, theta #trace
        d_theta = derivative(theta, data, y_hat)
        theta_next(theta, d_theta, alpha) # update theta values
        y_hat = logistic_y_hat(theta, data) # estimated output values
        error_prev = error # set previous error to current error
        error = logistic_error(data, y_hat) # calculate error
    return theta
```
